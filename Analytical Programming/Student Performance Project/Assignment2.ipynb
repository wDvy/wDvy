```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('StudentsPerformance.csv')
```

```python
#Part 2

scoreavgs = pd.Series()

for x in range(len(data)):
    savg = np.average(data.iloc[x:x+1,5:])
    scoreavgs[x] = savg

data["Score Averages"] = round(scoreavgs, 2)
data
```

```python

grades = ['A','B','C','D']
thresholds = [90,80,70,60]

gd = pd.Series()


for x in range(len(data)):
    if data["Score Averages"][x] >= 90: gd[x] = 'A'
    elif data["Score Averages"][x] >= 80: gd[x] = 'B'
    elif data["Score Averages"][x] >= 70: gd[x] = 'C'
    elif data["Score Averages"][x] >= 60: gd[x] = 'D'
    else: gd[x] = 'F'
    
data["Grades"] = gd

malegroup = data.query('gender == "male"')
femalegroup = data.query('gender == "female"')

print("Male Group")
print(malegroup.groupby('Grades').count()['gender'])

print()

print("Female Group")
print(femalegroup.groupby('Grades').count()['gender'])
```
